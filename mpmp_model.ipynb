```python
import random as rd
import pulp as pl
from pulp import *
rd.seed(1000)
P = [i+1 for i in range(5)]
R = [i+1 for i in range(3)]
T = [i+1 for i  in range(5)]
```

```python
demand = {}
for p in P:
  for t in T:
    demand.update({(p,t):rd.randint(1,100)})
print(demand)

usage = {}
for r in R:
  for p in P:
    usage.update({(r,p):rd.randint(1,10)})
print(usage)

capacity = {}
for r in R:
  capacity.update({r:rd.randint(800,1000)})
print(capacity)

inventory_cost = {}
for p in P:
  inventory_cost.update({p:rd.randint(1,10)})
print(inventory_cost)

production_cost = {}
for p in P:
  production_cost.update({p:rd.randint(1,10)})
print(production_cost)

shortage_cost = {}
for p in P:
  shortage_cost.update({p:rd.randint(1,10)})
print(shortage_cost)



```

```python
prob1 = pl.LpProblem("Production_Planning_Problem", pl.LpMinimize)
```

```python
#decision variables

I = {}  
for p in P:
    for t in T:
        I[p, t] = pulp.LpVariable("inv_" + str(p) + '_' + str(t), cat='Integer', lowBound=0)
print(I)

S = {}
for p in P:
    for t in T:
        S[p, t] = pulp.LpVariable("shor_" + str(p) + '_' + str(t), cat='Integer',lowBound=0)
print(S)

X = {}
for p in P:
    for t in T:
        X[p, t] = pulp.LpVariable("prod_" + str(p) + '_' + str(t), cat='Integer',lowBound=0)

print(X)

```

```python
#objective function
prob1 += pl.lpSum(inventory_cost[p] * I[p,t] + production_cost[p] * X[p,t] + shortage_cost[p] * S[p,t] for p in P for t in T)

print(prob1)
```

```python
#constraints
for p in P:
  for j in [0]:
    I[p,j] = 100

for p in P:
  for t in T:
    prob1 += I[p,t] == I[p,t-1]+demand[p,t]-X[p,t]+S[p,t] 
    for p in P:
      prob1 += usage[r,p]*X[p,t] <= capacity[r] 
print(prob1)
   

    
```

```python
status = prob1.solve()
optimization_result = prob1.solve()
print("Status:", LpStatus[prob1.status])
# print("Optimal Solution to the problem: ", value(prob.objective))
print ("Individual decision_variables: ")
for v in prob1.variables():
	print(v.name, "=", v.varValue)
```
